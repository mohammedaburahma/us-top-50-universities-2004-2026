## EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Research_Impact_Score'], bins=10, kde=True)
plt.title(f'{plot_no}. Distribution of Research Impact Scores Across Top Universities')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Employment_Rate'], bins=10, kde=True)
plt.title(f'{plot_no}. Employment Rate Distribution and Graduate Outcome Strength')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Research_Impact_Score', y='Employment_Rate', hue='Institution_Type')
plt.title(f'{plot_no}. Research Impact vs Employment Rate by Institution Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Intl_Student_Ratio', y='Research_Impact_Score', hue='Institution_Type')
plt.title(f'{plot_no}. International Student Ratio vs Research Impact')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='National_Rank', y='Research_Impact_Score')
plt.gca().invert_xaxis()
plt.title(f'{plot_no}. Relationship Between National Rank and Research Excellence')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='National_Rank', y='Employment_Rate')
plt.gca().invert_xaxis()
plt.title(f'{plot_no}. National Rank vs Graduate Employment Outcomes')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Institution_Type', y='Research_Impact_Score')
plt.title(f'{plot_no}. Research Impact Comparison: Public vs Private Institutions')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Institution_Type', y='Employment_Rate')
plt.title(f'{plot_no}. Employment Rate Comparison by Institution Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Institution_Type')
plt.title(f'{plot_no}. Distribution of Universities by Institution Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Founded_Year'], bins=15, kde=True)
plt.title(f'{plot_no}. Distribution of University Founding Years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Founded_Year', y='Research_Impact_Score')
plt.title(f'{plot_no}. Founding Year vs Research Impact Trend')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Founded_Year', y='Employment_Rate')
plt.title(f'{plot_no}. Founding Year vs Employment Success Rate')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df[['National_Rank','Founded_Year','Research_Impact_Score','Intl_Student_Ratio','Employment_Rate']].corr(), annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Matrix of Key University Performance Indicators')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_states = df['State'].value_counts().head(10)
sns.barplot(x=top_states.index, y=top_states.values)
plt.title(f'{plot_no}. Top 10 States by Number of Ranked Universities')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
state_research = df.groupby('State')['Research_Impact_Score'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=state_research.index, y=state_research.values)
plt.title(f'{plot_no}. Top 10 States by Average Research Impact')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Intl_Student_Ratio', y='Employment_Rate')
plt.title(f'{plot_no}. International Diversity vs Employment Outcomes')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(data=df, x='Research_Impact_Score', y='Employment_Rate')
plt.title(f'{plot_no}. Regression Trend: Research Strength Predicting Employment Rate')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Institution_Type', y='Intl_Student_Ratio')
plt.title(f'{plot_no}. International Student Ratio Distribution by Institution Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.kdeplot(df['Research_Impact_Score'], fill=True)
sns.kdeplot(df['Employment_Rate'], fill=True)
plt.title(f'{plot_no}. Density Comparison: Research Impact vs Employment Rate')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top10 = df.nsmallest(10, 'National_Rank')
sns.barplot(data=top10, x='University_Name', y='Employment_Rate')
plt.xticks(rotation=90)
plt.title(f'{plot_no}. Employment Rate Among Top 10 Nationally Ranked Universities')
show_fig()
plot_no += 1
```
